```python
import os

train_dir = "/content/drive/MyDrive/brats7d/data/BraTS2020_TrainingData/MICCAI_BraTS2020_TrainingData"
val_dir = "/content/drive/MyDrive/brats7d/data/BraTS2020_ValidationData/MICCAI_BraTS2020_ValidationData"

missing = []
for case in os.listdir(train_dir):
    seg_path = os.path.join(train_dir, case, f"{case}_seg.nii")
    if not os.path.exists(seg_path):
        missing.append(seg_path)

print("Missing segmentations:", missing[:10])  # show first 10
print("Total missing:", len(missing))
```
